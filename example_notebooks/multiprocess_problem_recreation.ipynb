```python
import os

import multiprocess as mp
import numpy as np
import scipy.linalg as linalg

np.random.seed(42)
```

```python
class MyClass:
    def fit(self, X, y):
        print(f"{os.getpid()}, started")
        res = linalg.solve(X, y, check_finite=False)
        print(f"{os.getpid()}, done")
        return res


def create_worker_fn():
    my_class = MyClass()

    def worker_fn(args):
        X, y = args
        return my_class.fit(X, y)

    return worker_fn


X = np.eye(10) + np.random.randn(10, 10) * 0.001
y = np.random.randn(10, 10)

worker_fn = create_worker_fn()

pool = mp.Pool(processes=10)
results = pool.map_async(worker_fn, [(X, y) for _ in range(10)])
pool.close()
pool.join()
print(results.get(timeout=10))
```

```python
for p in range(1, 10):
    worker_fn = create_worker_fn(p)
    worker_fn(X, y)
```
